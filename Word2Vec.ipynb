# Word2Vec is the technique which convets words into vector like bag of words and tf-idf but the greatest advantanges of Word2Vec is that it provide the semantic meaning of the words. for example: happy, joy are the two words which are similar which are captured by word2vec but could not told by bag of words and tf-idf. Word2Vec create low dimensional vector and dense vector while bag of words,tf-idf create very high dimensional vector and sparse vector

```python
import numpy as np
import pandas as pd
import gensim
import os
from gensim.models import Word2Vec,KeyedVectors
import nltk
nltk.download('punkt')
```

```python
import os
from nltk import sent_tokenize, word_tokenize
from gensim.utils import simple_preprocess
from nltk.corpus import stopwords

# Download stopwords (if not already downloaded)
import nltk
nltk.download('stopwords')

# Load English stopwords from NLTK
stop_words = set(stopwords.words('english'))

story = []

# Specify the correct directory path
directory_path = '/content/drive/MyDrive/Gameofthrones'

for filename in os.listdir(directory_path):
    file_path = os.path.join(directory_path, filename)

    with open(file_path, 'r', encoding='latin-1') as file:
        corpus = file.read()
        raw_sentences = sent_tokenize(corpus)

        for sent in raw_sentences:
            # Tokenize the sentence into words
            words = simple_preprocess(sent)

            # Remove stopwords
            filtered_words = [word for word in words if word.lower() not in stop_words]

            story.append(filtered_words)
```

```python
len(story)
```

```python
model = gensim.models.Word2Vec(
    window =7,
    min_count=2,
)
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=5)
```
