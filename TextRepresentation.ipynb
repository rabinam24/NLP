# Here we will do some Text Representation which is also known as feature extraction

# Text representation is converting the text into the numbers with the meaningful value, that's the meaning of text must be conserved while converting into the numbers

# First technique is the one hot encoding where we convert the text into corpus which meaning concatenating all the words, and defining the vocabulary which means identify the unique words present in the corpus and then we convert the text into the number which you can google and understand more deeply. The problem with this is that it create the sparsity and lost the meaning while converting into the numbers.

# Bag of words

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'text':['people watch football','football watch football','people support team','football support team'],'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer()
```

```python
bow =cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(['People support and watch football for the happiness of the people']).toarray()
```

## Here above, we can see it is handeling out of the vocabulary problem easily like for the happiness of! but still it's has some problem considering the out of the vocabulary, sparsity,non-ordering of the words can create different meaning and can create problem and finally, for example: I like watching football, I don't like watching football, in this case since there is one word only difference, bag of words will tries to define the less difference in them with less angle. but They are completely different which can be solve by now ngrams.

# ngrams bigrams

```python
df1 = pd.DataFrame({'text':['people watch football','football watch football','people support team','football support team'],'output':[1,1,0,0]})
```

```python
df1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(ngram_range=(2,2))
```

```python
bow =cv.fit_transform(df1['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

# Tf-Idf (Term frequency-Inverse document frequency) concept is that it assign different values to different words present in the document based on tf-idf values

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf.fit_transform(df['text']).toarray()
```

```python
print(tfidf.idf_)
print(tfidf.get_feature_names_out())
```
